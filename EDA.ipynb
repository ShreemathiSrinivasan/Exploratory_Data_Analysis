```python
#seperation of numerical and categorical data for easy analysis
cat_cols=df.select_dtypes(include=['object']).columns
num_cols = df.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
```

```python
sns.countplot(data=df,x='Project Type')
plt.xticks(rotation=45)
plt.title('Project Count by Type')
plt.show()
```

```python
sns.barplot(data=df,x='Department',y=' Project Cost ',estimator=np.mean)
plt.xticks(rotation=45)
plt.title('Average Cost by Department')
plt.show()
```

```python
plt.pie(df['Status'].value_counts(), labels=df['Status'].value_counts().index,autopct='%1.1f%%')
plt.title('Project Status Distribution')
plt.show()
```

```python
corr=df.corr()
sns.heatmap(corr, annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#boxplot
sns.boxplot(data=df, x='term', y='loan_amount',showfliers=False)
plt.title('Loan Term by Loan Amount')
plt.show()
```

```python
sns.boxplot(data=df,x='Complexity',y='Completion%',showfliers=False)
plt.title('Project Completion% by Complexity')
plt.show()
```

```python
df['Year-Month']=df['Year'].astype(str)+'-'+df['Month'].astype(str)
df.groupby('Year-Month')[' Project Cost '].sum().plot()
plt.title('Monthly Project Costs Over Time')
plt.xticks(rotation=45)
plt.show()
```

```python
#linechart
df['Categorical_Var'] = df['home_ownership'] + '-' + df['purpose']
df.groupby('Categorical_Var')['loan_amount'].sum().plot()
plt.title('Sum of Loan Amounts by Categorical Variable')
plt.xticks(rotation=45)
plt.show()
```

```python
#histogram
sns.histplot(data=df, x='loan_amount', kde=True)
plt.title("Loan Amounts")
plt.show()
```

```python
#countplot
sns.countplot(data=df, x="interest_payments")
plt.xticks(rotation=45)
plt.title("Interest Payments Count by type")
plt.show()
```

```python
#barplot
sns.barplot(data=df, x="region", y="loan_amount", estimator= np.mean)
plt.xticks(rotation=45)
plt.title('Average loan By Region')
plt.show()
```

```python
#piechart
plt.pie(df['region'].value_counts(), labels=df['region'].value_counts().index,autopct='%1.1f%%')
plt.title('Region Distribution')
plt.show()
```

```python
plt.pie(df['income_category'].value_counts(),labels=df['income_category'].value_counts().index,autopct='%1.1f%%')
plt.title('Income Category Distribution')
plt.show()
```

```python
variables_of_interest = ['annual_inc', 'loan_amount', 'interest_rate', 'dti']
subset_df = df[variables_of_interest]

corr = subset_df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df['purpose']=df['purpose']
df.groupby('purpose')['loan_amount'].sum().plot()
plt.title('Purpose and Loan Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
import plotly.express as px
fig = px.scatter_3d(df, x='annual_inc', y='loan_amount', z='term',color='application_type')
fig.update_layout(scene = dict(
                    xaxis = dict(title  = 'Annual Income'),
                    yaxis = dict(title  = 'Loan Amount'),
                    zaxis = dict(title  = 'Term')))
fig.show()
```

```python
sns.scatterplot(data=df, x='loan_amount',y='annual_inc')
plt.title("loan amount and term")
plt.show()
```

```python
df[' Project Cost ']=df[' Project Cost '].str.replace(',','').astype(float)
```

```python
sns.scatterplot(data=df,x=' Project Cost ',y=' Project Benefit ')
plt.title(' Project Cost vs. Benefit ')
plt.show()
```
